Task - 3 - Outlier Detection:

Tasks:

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 Ã— IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
# 1 and 2: Identify outlier countries in South Asia based on their Score and GDP per Capita.
#          Define outliers using the 1.5 Ã— IQR rule.
data=pd.read_csv('/content/south_asia_happiness.csv')
def find_outliers(column):
    q1 = np.percentile(data[column], 25)
    q3 = np.percentile(data[column], 75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return lower_bound, upper_bound
outlier_score = find_outliers('score')
outlier_gdp = find_outliers('Log GDP per capita')

data['Outlier'] = (
    (data['score'] < outlier_score[0]) | (data['score'] > outlier_score[1]) |
    (data['Log GDP per capita'] < outlier_gdp[0]) | (data['Log GDP per capita'] > outlier_gdp[1])
)
outliers = data[data['Outlier']]
print("Outlier Countries: ")
print(outliers[['Country name','score','Log GDP per capita','Outlier']])
print("\n")
```

```python
# 3: Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis,
#    highlighting outliers in a different color
plt.figure(figsize=(10, 6))
plt.scatter(data['Log GDP per capita'], data['score'], label='Non-Outliers', color='blue', alpha=0.7)
plt.scatter(outliers['Log GDP per capita'], outliers['score'], label='Outliers', color='red', edgecolor='black',s=100)
plt.xlabel('Log GDP per Capita', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)
plt.title(' GDP vs Happiness Score of South Asian Countries', fontsize=14)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

Task - 4 - Exploring Trends Across Metrics:

Tasks:
1. Choose two metrics (e.g., Freedom to Make Life Choices and Generosity) and calculate their correlation
{pearson correlation} with the Score for South Asian countries.
2. Create scatter plots with trendlines for these metrics against the Score.
3. Identify and discuss the strongest and weakest relationships between these metrics and the Score for
South Asian countries.

```python
# 1: Choose two metrics (e.g., Freedom to Make Life Choices and Generosity)
#    and calculate their correlation {pearson correlation} with the Score for South Asian countries.
data = pd.read_csv('/content/south_asia_happiness.csv')

metrics = ['Freedom to make life choices', 'Generosity']
correlations = {}

for metric in metrics:
    corr = data['score'].corr(data[metric])
    correlations[metric] = corr

print("Pearson Correlations with Happiness Score:")
for metric, corr in correlations.items():
    print(f"{metric}: {corr:.2f}")
```

```python
# 2: Create scatter plots with trendlines for these metrics against the Score.
def scatter_with_trendline(x, y, xlabel, ylabel, title):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    trendline = slope * x + intercept


    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, label='Data Points', color='blue', alpha=0.7)
    plt.plot(x, trendline, label=f'Trendline (RÂ² = {r_value**2:.2f})', color='red')
    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)
    plt.title(title, fontsize=14)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()
for metric in metrics:
    scatter_with_trendline(x=data[metric],y=data['score'],xlabel=metric,ylabel='Happiness Score',title=f'Happiness Score vs {metric}')
```

```python
# 3: Identify and discuss the strongest and weakest relationships
#    between these metrics and the Score for South Asian countries.
strongest_metric = max(correlations, key=correlations.get)
weakest_metric = min(correlations, key=correlations.get)

print("\nAnalysis:")
print(f"The strongest relationship is with '{strongest_metric}' (Correlation: {correlations[strongest_metric]:.2f}).")
print(f"The weakest relationship is with '{weakest_metric}' (Correlation: {correlations[weakest_metric]:.2f}).")
```

6. Outlier Detection:

  â€¢ Identify outlier countries in both regions based on Score and GDP per Capita.

  â€¢ Plot these outliers and discuss their implications.

```python
# Identify outlier countries in both regions based on Score and GDP per Capita.
def detect_outliers(column, region_data):
    Q1 = region_data[column].quantile(0.25)
    Q3 = region_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return region_data[(region_data[column] < lower_bound) | (region_data[column] > upper_bound)]

south_asia_outliers = detect_outliers('score', south_asia_data)
middle_east_outliers = detect_outliers('score', middle_east_data)

print("Outlier countries in Souch Asia:\n")
print(south_asia_outliers[['Country name']])
print("\nOutlier countries in Middle East:\n")
print(middle_east_outliers[['Country name']])
```
